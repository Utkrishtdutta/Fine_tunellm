## Post-Fine Tuning(GPT-2)

We are passing the question to `Fine-tuned GPT-2` which we have fine tuned using our data.

```python
import pandas as pd
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

Here, we are declaring a function that is taking question as prompt_text and returning text-generated from `Fine-tuned GPT-2` as output. For this we are using `Tokenizer` we have saved during training and `model` to get response from Fine-tuned GPT-2.

```python
def fine_tune_gpt_2(prompt_text):
    try:
        model = GPT2LMHeadModel.from_pretrained("fine-tuned-gpt2")
        tokenizer = GPT2Tokenizer.from_pretrained("fine-tuned-gpt2")
        input_ids = tokenizer.encode(prompt_text, return_tensors="pt")
        max_length = min(len(prompt_text.split(' '))*20,250)
        output = model.generate(input_ids, num_return_sequences=1,max_length=max_length, temperature=0, pad_token_id=tokenizer.eos_token_id, attention_mask=torch.ones(input_ids.shape))
        generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
        return generated_text.split('[SEP]')[1]
    except Exception:
        pass
```

We are passing question one-by-one to above function after initializing dataframe from excel file.

```python
df = pd.read_excel(r'training_data_assignment.xlsx')
df.columns = ['question', 'answer']
df['answer_posttrain'] = df['question'].apply(lambda x : fine_tune_gpt_2(x))
```

We are saving the response generated by LLM in excel sheet.

```python
df.to_excel('post_train_response.xlsx',index=False)
```
