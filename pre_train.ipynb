## Pre-Fine Tuning(GPT-2)

We are passing the question to `GPT-2` which we have imported from Hugging Face using it's library transformer.

```python
import pandas as pd
from transformers import pipeline,AutoTokenizer,AutoModelForCausalLM
```

Here, we are declaring a function that is taking question as argument and returning text-generated from `GPT-2` as output. For this we are using `AutoTokenizer` for tokenization and `AutoModelForCausalLM` to get response from GPT-2.

```python
def gpt_2_model(description):
    """
    Function to get text description through GPT2.
    """
    tokenizer = AutoTokenizer.from_pretrained('gpt2')

    pipe = pipeline("text-generation", model=AutoModelForCausalLM.from_pretrained('gpt2'), tokenizer=tokenizer)
    response = pipe(description, min_length=1)
    return response[0]['generated_text']
```

We are passing question one-by-one to above function after initializing dataframe from excel file.

```python
df = pd.read_excel(r'training_data_assignment.xlsx')
df.columns = ['question', 'answer']
df['answer_pretrain'] = df['question'].apply(lambda x : gpt_2_model(x))
```

We are saving the response generated by LLM in excel sheet.

```python
df.to_excel('pre_train_response.xlsx',index=False)
```
